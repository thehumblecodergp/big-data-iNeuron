Two major datatypes
1. Series
2. DataFrame

```python
lst = [1,2,3,4]
print(type(lst))
```

```python
# Convert this list to series.
ser = pd.Series(lst)
print(type(ser))
ser

df = pd.DataFrame(lst)
print(type(df))
df
```

Series is a one dimensional datatype.

```python
# Changing the multi-dimensional list to series.
lst1 = [[1,2,3,4],[5,6,7,8]]
ser1 = pd.Series(lst1)
ser1
```

DataFrame is a datatype having more than 1 dimension.

Number of columns is considered as dimension. If the dataset has one coloumn we can consider the dataset as series else it will be dataframe.

```python
# Changing the multi-dimesional list to dataframe
df1 = pd.DataFrame(lst1)
df1
```

```python
# GroupBy Operation
data1 = {
    'Name': ['Jai','Anuj','Jai','Prince','Gourav','Anuj','Prince','Abhi'],
    'Age':[27,24,25,28,29,30,29,31],
    'Address': ['Kanpur','Nagpur','Allahabad','Kannauj','Jaunpur','Kanpur','Allahabad','Nagpur'],
    'Qualifiction': ['MCA','MSC','MS','M.Tech','B.Tech','MBA','BSc','B.Tech'],
}
```

```python
type(data1)
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
# Group-by the data by name
df1.groupby('Name')
print(df1.groupby('Name').groups)
```

```python
# Gouping all the information on the basis of a column
gk = df1.groupby('Name')
lst = list(gk)
lst
```

```python
for i in gk:
    print(i)
```

```python
# Group by using multiple keys
print(df1.groupby(['Name','Qualifiction']).groups)
```

```python
df1.groupby(['Name']).sum()
```

```python
# Mering two dataframes
data1 = {
    'Name': ['Jai','Anuj','Jai','Prince'],
    'Age':[27,24,25,28],
    'Address': ['Kanpur','Nagpur','Allahabad','Kannauj'],
    'Qualifiction': ['MCA','MSC','MS','M.Tech'],
}

df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
    'Name': ['Gourav','Anuj','Prince','Abhi'],
    'Age':[29,30,29,31],
    'Address': ['Jaunpur','Kanpur','Allahabad','Nagpur'],
    'Qualifiction': ['B.Tech','MBA','BSc','B.Tech'],
}
df2 = pd.DataFrame(data2)
df2
```

```python
df_merged = [df1, df2]
res = pd.concat(df_merged)
res
```
